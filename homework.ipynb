### Problem 10

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]
````

````python
# YOUR CHANGES HERE

# get the last element of q10
# last_element = q10[-1]
# let x = last_element
for i in range(10):
    # get the last element of q10
    last_element = q10[-1]
    x = last_element

    # for odd x-values
    if x % 2 == 1:
        new_x_value = 3*x + 1
        q10.append(new_x_value)
        
    # for even x-values
    else:
        new_x_value = x / 2
        q10.append(new_x_value)

print(q10)
````

### Problem 15

Use the function `math.comb` to make a bar chart illustrating the binomial distribution for $n=10$.
That is, for integers x from zero to ten, there should be a bar whose height is $10 \choose x$.

````python
# YOUR CHANGES HERE

# use plt.bar() to create bar chart
for x in range(11):
    # height (y) of each bar is "10 choose x"
    y = math.comb(10, x)
    plt.bar(x, y)

    # add title and labels
    plt.title("Binomial Distribution")
    plt.xlabel("x")
    plt.ylabel("y (10 choose x)")
````

### Problem 17

Plot the factorial function from n=1 to 10 using log scale for better legibility.

Recall that the factorial of $n$, denoted $n!$, is calculated as the product $1 \times 2 \times \ldots \times n$.
The plots that you plot should be $(1, 1!), (2, 2!), (3, 3!), \ldots, (10, 10!)$.
Use [`plt.yscale`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html) to change your y axis to use log scale.

````python
# YOUR CHANGES HERE

# create list of x-values from 1 to 10, excluding 0
x_values = [i for i in range(11) if i > 0]
#print(x_values)

# create list of y-values that are factorials of x-values
y_values = [math.factorial(i) for i in x_values]
#print(y_values)

# plot x and y values using plt.yscale() to set y-axis to logarithmic scale
plt.plot(x_values, y_values)
plt.yscale("log")

# add title and labels
plt.title("Some Factorial Function")
plt.xlabel("x (in range 1-10)")
plt.ylabel("y (log scale)")
````

The file `abalone.tsv` has data from the [UCI Abalone data set](https://archive.ics.uci.edu/dataset/1/abalone).
Load the file and plot it as a scatter plot using the length for the x axis and diameter for the y axis.
Make sure to label the chart.

````python
# YOUR CHANGES HERE

import csv

# create empty lists to store x and y values
lengths = []
diameters = []

# parse abalone.tsv data
with open("abalone.tsv", newline="") as file:
    reader = csv.reader(file, delimiter="\t")
    header = next(reader)  # Skip header row 

    # find column indices
    length_idx = header.index('Length')
    diameter_idx = header.index('Diameter')

    for row in reader:
        lengths.append(float(row[length_idx]))
        diameters.append(float(row[diameter_idx]))
        
# plot points using plt.scatter()
plt.scatter(lengths, diameters, s=5)
        
# add title and labels
plt.title("Abalone Diameter vs Length")
plt.xlabel("Length")
plt.ylabel("Diameter")
````
